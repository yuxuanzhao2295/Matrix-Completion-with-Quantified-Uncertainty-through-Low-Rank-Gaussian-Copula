```julia
# Continuous with l2
resTuning_cont = zeros(20,9,2,2);
R"sigmaSeq = c(0.1,0.5)";
R"ratio = exp(seq(from=log(1/4),to=log(1/100),length=9))"
for i = 1:20
    for j = 1:2
        @rput i j;
        R"if (j==1) setting = 'LR' else setting = 'HR'"
        R"Xtrue = generate_LRGC(n=500, p=c(200,0,0), rank = 10, sigma = 0.1, seed = i, cont.type = setting)";
        R"m = mask(Xtrue,0.4,seed=i)";
        R"loc_obs = m$loc_obs";
        R"loc_mis = m$loc_mis";
        @rget Xtrue loc_obs loc_mis;
        
        R"xc = biScale(m$Xobs,row.scale = FALSE,col.scale = FALSE)"
        R"lam0=lambda0(xc)"
        @rget lam0 ratio;
        for s = 1:9
            obj = imp_MMMF_l2(lam0 * ratio[s],loc_obs, loc_mis, Xtrue);
            resTuning_cont[i,s,1,j] = obj[1]
            resTuning_cont[i,s,2,j] = obj[2]
        end
    end
end
```

```julia
@rput resTuning_cont;
R"dimnames(resTuning_cont) = list(NULL, NULL, c('NRMSE','time'), c('LR','HR'))";
R"err = apply(resTuning_cont, c(2,3,4), mean)";
R"err[,1,]"
```

```julia
# Ordinal with BvS and l1
resTuning_ord = zeros(20,9,2,2,2);
R"sigmaSeq = c(0.1,0.5)";
R"ratio = exp(seq(from=log(1/4),to=log(1/100),length=9))"
for i = 1:20
    for j = 1:2
        @rput i j;
        R"Xtrue = generate_LRGC(n=500, p=c(0,200,0), rank = 5, sigma = sigmaSeq[j], seed = i)";
        R"m = mask(Xtrue,0.6,seed=i)";
        R"loc_obs = m$loc_obs";
        R"loc_mis = m$loc_mis";
        @rget Xtrue loc_obs loc_mis;
        Xtrue = convert(Array{Int64,2}, Xtrue);
        
        R"xc = biScale(m$Xobs,row.scale = FALSE,col.scale = FALSE)"
        R"lam0=lambda0(xc)"
        @rget lam0 ratio;
        for s = 1:9
            obj = imp_MMMF_BvS(lam0 * ratio[s],loc_obs, loc_mis, Xtrue);
            resTuning_ord[i,s,1,j,1] = obj[1]
            resTuning_ord[i,s,2,j,1] = obj[2]
            obj = imp_MMMF_l1(lam0 * ratio[s],loc_obs, loc_mis, Xtrue);
            resTuning_ord[i,s,1,j,2] = obj[1]
            resTuning_ord[i,s,2,j,2] = obj[2]
        end
    end
end
```

```julia
@rput resTuning_ord;
R"dimnames(resTuning_ord) = list(NULL, NULL, c('MAE','time'), c('HighSNR','LowSNR'), c('BvS','l1'))";
R"err = apply(resTuning_ord, c(2,3,4,5), mean)";
R"err[,1,,]"
```

```julia
# Binary with Hinge and Logistic
resTuning_bin = zeros(20,9,2,2,2);
R"sigmaSeq = c(0.1,0.5)";
R"ratio = exp(seq(from=log(1),to=log(1/100),length=9))"
for i = 1:20
    for j = 1:2
        @rput i j;
        R"Xtrue = generate_LRGC(n=500, p=c(0,0,200), rank = 5, sigma = sigmaSeq[j], seed = i)";
        R"m = mask(Xtrue,0.6,seed=i)";
        R"loc_obs = m$loc_obs";
        R"loc_mis = m$loc_mis";
        @rget Xtrue loc_obs loc_mis;
        Xtrue = [x==1 for x in Xtrue];
        
        R"xc = biScale(m$Xobs,row.scale = FALSE,col.scale = FALSE)"
        R"lam0=lambda0(xc)"
        @rget lam0 ratio;
        for s = 1:9
            obj = imp_MMMF_binary(lam0 * ratio[s],loc_obs, loc_mis, Xtrue,LogisticLoss());
            resTuning_bin[i,s,1,j,1] = obj[1]
            resTuning_bin[i,s,2,j,1] = obj[2]
            obj = imp_MMMF_binary(lam0 * ratio[s],loc_obs, loc_mis, Xtrue,HingeLoss());
            resTuning_bin[i,s,1,j,2] = obj[1]
            resTuning_bin[i,s,2,j,2] = obj[2]
        end
    end
end
```

```julia
@rput resTuning_bin;
R"dimnames(resTuning_bin) = list(NULL, NULL, c('MAE','time'), c('HighSNR','LowSNR'), c('logistic','hinge'))";
R"err = apply(resTuning_bin, c(2,3,4,5), mean)";
R"err[,1,,]"
```

```julia
# Continuous error by reliability
loss = QuadLoss();
R"ratio = exp(seq(from=log(1/4),to=log(1/100),length=9))[c(4,6)]"
R"errReliability_cont_MMMF = array(0, dim = c(20,100,2))"
@rget ratio;
for i = 1:20
    for j = 1:2
        @rput i j;
        
        R"if (j==1) setting = 'LR' else setting = 'HR'"
        R"Xtrue = generate_LRGC(n=500, p=c(200,0,0), rank = 10, sigma = 0.1, seed = i, cont.type = setting)";
        R"m = mask(Xtrue,0.4,seed=i)";
        R"loc_obs = m$loc_obs";
        R"loc_mis = m$loc_mis";
        R"xc = biScale(m$Xobs,row.scale = FALSE,col.scale = FALSE)"
        R"lam0=lambda0(xc)"
        @rget Xtrue loc_obs loc_mis lam0;
        
        output = imp_MMMF(loss,lam0*ratio[j], loc_obs, loc_mis, Xtrue, false, 199);
        ximp = output[:,1]; xtrue = output[:,2];
        
        R"mm = CV_split(m$Xobs, nfold = 10, seed = i)"
        Ximp = zeros(length(ximp),10);
        for s = 1:10
            @rput s;
            R"loc_obs_new = mm$obs.2d[[s]]"
            @rget loc_obs_new;
            
            R"Xobs = Xtrue; Xobs[mm$mis.1d[[s]]] = NA"
            R"xc = biScale(Xobs,row.scale = FALSE,col.scale = FALSE)"
            R"lam0=lambda0(xc)"
            @rget lam0;
            
            Ximp[:,s] = imp_MMMF_XY(loss,lam0*ratio[j], loc_obs_new, loc_mis, Xtrue, 199);
        end
        r = var(Ximp, dims=2);
        @rput r ximp xtrue
        R"errReliability_cont_MMMF[i,,j] = nrmse_by_reliability(r = -r, xtrue = xtrue, ximp = ximp)";
    end
end
```

```julia
R"err = apply(errReliability_cont_MMMF, c(2,3), mean)";
R"plot(err[,1])"
```
