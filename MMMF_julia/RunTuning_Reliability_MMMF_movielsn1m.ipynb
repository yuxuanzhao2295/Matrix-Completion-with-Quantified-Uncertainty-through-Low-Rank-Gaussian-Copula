```julia
R"movie_50obs = read.csv('movielens_50obs.csv')";
R"movie_50obs = as.matrix(movie_50obs)";
@rget movie_50obs;
```

```julia
function eval_split_bvs_GLRM(data, lambda, loc_train, loc_test, loc_val)
    # 
    l_test, l_val, l_train = size(loc_test)[1], size(loc_val)[1], size(loc_train)[1];
    loc_test = [(loc_test[j,1], loc_test[j,2]) for j in 1:l_test];
    loc_val = [(loc_val[j,1], loc_val[j,2]) for j in 1:l_val];
    loc_train = [(loc_train[j,1], loc_train[j,2]) for j in 1:l_train];
    #
    ry = QuadReg(lambda);
    rx = QuadReg(lambda);
    k=200
    glrm = GLRM(data, BvSLoss(5),rx,ry,k, obs = loc_train);
    add_offset!(glrm)
    init_svd!(glrm);
    @timed begin
        fit!(glrm, ProxGradParams(max_iter = 500));
        imp = impute_missing(glrm);
        Val_test = convert(Array{Int8}, [data[i[1], i[2]] for i in loc_test]);
        Imp_test = convert(Array{Int8}, [imp[i[1], i[2]] for i in loc_test]);
        Val_val = convert(Array{Int8}, [data[i[1], i[2]] for i in loc_val]);
        Imp_val = convert(Array{Int8}, [imp[i[1], i[2]] for i in loc_val]);
        err = zeros(4);
        err[1] = mean(abs.(Val_test - Imp_test));
        err[2] = mean((Val_test - Imp_test).^2)^.5;
        err[3] = mean(abs.(Val_val - Imp_val));
        err[4] = mean((Val_val - Imp_val).^2)^.5;
        err
    end
end
```

```julia
info_movie50obs_MMMF = zeros(5,4,5);
ratio = [1/10, 1/12, 1/14, 1/16];
for i in 1:5
    @rput i;
    R"m = data_split(movie_50obs,seed = i)";
    R"loc_test = m$loc_test";
    R"loc_val = m$loc_val";
    R"loc_train = m$loc_train";
    R"Xtrain = m$Xtrain"
    R"xc = biScale(Xtrain,row.scale = FALSE,col.scale = FALSE)"
    R"lam0=lambda0(xc)"
    @rget lam0 loc_test loc_val loc_train Xtrain;
    @rget seq;
    for j in 1:4
        obj = eval_split_bvs_GLRM(movie_50obs, lam0*ratio[j], loc_train, loc_test, loc_val);
        info_movie50obs_MMMF[i,j,1:4] = obj[1]
        info_movie50obs_MMMF[i,j,5] = obj[2]
    end
end
```

```julia
# error by quantile
loss = BvSLoss(5);
ratio = 1/14;
R"errReliability_movielens_MMMF = array(0, dim = c(5,100,2))"
for i = 1:5
    @rput i;
    
    R"m = data_split(movie_50obs,seed = i)";
    R"loc_test = m$loc_test";
    R"loc_val = m$loc_val";
    R"loc_train = m$loc_train";
    R"Xtrain = m$Xtrain"
    R"xc = biScale(Xtrain,row.scale = FALSE,col.scale = FALSE)"
    R"lam0=lambda0(xc)"
    @rget lam0 loc_test loc_val loc_train Xtrain;
    
    output = imp_MMMF(loss, lam0*ratio, loc_train, loc_test, movie_50obs, true, 200, 500)
    ximp = output[:,1]; xtrue = output[:,2];

    R"mm = CV_split(m$Xtrain, nfold = 10, seed = i)"
    Ximp = zeros(length(ximp),10);
    for s = 1:10
        @rput s;
        R"loc_obs_new = mm$obs.2d[[s]]"
        @rget loc_obs_new;
            
        R"Xobs = movie_50obs; Xobs[mm$mis.1d[[s]]] = NA"
        R"xc = biScale(Xobs,row.scale = FALSE,col.scale = FALSE)"
        R"lam0=lambda0(xc)"
        @rget lam0;
        
        Ximp[:,s] = imp_MMMF_XY(loss, lam0*ratio, loc_obs_new, loc_test, movie_50obs, 200, 500);
    end
    r = var(Ximp, dims=2);
    @rput r ximp xtrue
    R"errReliability_movielens_MMMF[i,,1] = mae_by_reliability(r = -r, xtrue = xtrue, ximp = ximp)";
    R"errReliability_movielens_MMMF[i,,2] = rmse_by_reliability(r = -r, xtrue = xtrue, ximp = ximp)";
end
```
